# Inner Join

```python
d = {
    "Id" : ["1", "2", "3", "4", "5", "6"],
    "Color" : ["Red", "Green", "Yellow", "Blue", "Pink", "Black"],
    "Fruit" : ["Apple", "Banana", "Mango", "Berry", "Melon", "Grapes"]}

d1=pd.DataFrame(d)
d1
```

```python
z = { "Id" : ["1", "2", "3", "4", "5", "7"],
     "Ratings" : ["A", "B", "C", "B", "A", "D"],
     "Cost" : [200, 300, 250, 400, 350, 150],
     "Fruit" : ["Apple", "Banana", "Mango", "Berry", "Melon", "Kiwi"],
     "Buy" : ["N", "Y", "Y", "N", "Y", "N"]}

d2 = pd.DataFrame(z)
d2
```

```python
join = pd.merge(d1, d2, on="Fruit", how="right")
join
```

```python
join = pd.merge(d1, d2, on="Fruit", how='left')
join
```

```python
join = pd.merge(d1[["Color", "Fruit"]], d2, on="Fruit", how="outer")

join
```

```python
df_merge = pd.merge(d1, d2, on=["Fruit", "Id"])

df_merge
```

```python
join = pd.merge(d1[["Color", "Fruit"]], d2, on="Fruit", how="inner")
join
```

```python
join = pd.merge(d1[["Color", "Fruit", "Id"]], d2, on="Id", how="inner")
join
```
